```python
#import the requerd libraries
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get("https://www.indeed.co.in/jobs?q=data+scientist&1=Delhi")
page
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
print(soup.prettify()) # now intentation is presented for understanding parenting and child tag
```

```python
# first we will extract the html tag where we have the first job tittle
first_title=soup.find('h2',class_='title') #it shows web page -->right click inspect
first_title
```

```python
first_title.text.replace('\n','')
```

```python
# we will extract the html tags where we have the first company names.
first_company=soup.find('span',class_="company")
first_company
```

```python
titles=soup.find_all('h2',class_="title")
titles
```

```python
job_titles=[]
for i in titles:
    job_titles.append(i.text)
job_titles    
```

```python
job_titles[3]
```

```python
job_titles=[]
for i in titles:
    job_titles.append(i.text.replace('\n',''))
job_titles    
```

```python
companies=soup.find_all('span',class_="company")
companies
```

```python
company_names=[]
for i in companies:
    company_names.append(i.get_text().replace('\n',''))
company_names    
```

```python
days =  soup.find_all('span',class_="date date-a11y")
days
```

```python
all_days=[]
for i in days:
    all_days.append(i.get_text().replace('\n',''))
all_days    
```

```python
locations=soup.find_all('span',class_="location accessible-contrast-color-location")
locations
```

```python
location_all=[]
for i in locations:
     location_all.append(i.get_text().replace('\n',''))
location_all        
    
```

```python
print(len(company_names),len(job_titles),len(all_days),len(location_all))
```
