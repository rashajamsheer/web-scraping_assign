```python
import selenium
import pandas as pd
import time
from selenium import webdriver
from bs4 import BeautifulSoup
from selenium.common.exceptions import StaleElementReferenceException,NoSuchElementException,ElementNotInteractableException,TimeoutException
import requests
import re
```

```python
driver = webdriver.Chrome("chromedriver.exe")
url = 'https://www.olx.in/cars_c84'
driver.get(url)
```

```python
urls=[]
for i in range(10):
    for j in driver.find_elements_by_xpath("//ul[@class='rl3f9 _3mXOU']/li/a"):
        urls.append(j.get_attribute("href"))
    driver.find_element_by_xpath("//button[@class='rui-39-wj rui-3evoE rui-1JPTg']|//div[@class='JbJAl']/button/span").click()
    time.sleep(3)
```

```python
len(urls)
```

```python
brand=[]
model=[]
variant=[]
year = []
km = []
fuel=[]
owner=[]
location=[]
price=[]
```

```python
for i in urls:
    driver.get(i)
    time.sleep(5)
    try:
        a = driver.find_element_by_xpath("//span[@data-aut-id='value_make']")
        brand.append(a.text)
    except NoSuchElementException:
        brand.append('NaN')
    
    try:
        a = driver.find_element_by_xpath("//span[@data-aut-id='value_model']")
        model.append(a.text)
    except NoSuchElementException:
        model.append('NaN')
        
    try:
        a = driver.find_element_by_xpath("//span[@data-aut-id='value_variant']")
        variant.append(a.text)
    except NoSuchElementException:
        variant.append('NaN')
        
    try:
        a = driver.find_element_by_xpath("//span[@data-aut-id='value_year']")
        year.append(a.text)
    except NoSuchElementException:
        year.append('NaN')
        
    try:
        a = driver.find_element_by_xpath("//span[@data-aut-id='value_mileage']")
        km.append(a.text)
    except NoSuchElementException:
        km.append('NaN')
        
        
    try:
        a = driver.find_element_by_xpath("//span[@data-aut-id='value_petrol']")
        fuel.append(a.text)
    except NoSuchElementException:
        fuel.append('NaN')
   
    try:
        a = driver.find_element_by_xpath("//span[@data-aut-id='value_first_owner']")
        owner.append(a.text)
    except NoSuchElementException:
        owner.append('NaN')
        
    try:
        a = driver.find_element_by_xpath("//ol[@class='rui-10Yqz']/li[5]/a")
        location.append(a.text)
    except NoSuchElementException:
        location.append('NaN')
        
    
    try:
        a = driver.find_element_by_xpath("//span[@data-aut-id='itemPrice']")
        price.append(a.text)
    except NoSuchElementException:
        price.append('NaN')
```

```python
df = pd.DataFrame({})
df["Brand"] = brand
df["Model"] = model
df["Variant"] = variant
df["Mileage"] = km
df["Fuel"] = fuel
df["Owner"] = owner
df["Location"] = location
df["Price"] = price
```

```python
df["Location"] = df["Location"].str.split(' in ',1).str[1]
```

```python
df.to_csv("olx.csv")
```
