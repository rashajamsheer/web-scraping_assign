# Q5)

Scrape the details of trending repositories on Github.com.


```python
driver = webdriver.Chrome("chromedriver.exe")
url = 'https://github.com/'
driver.get(url)
```

```python
driver.find_element_by_xpath('/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/summary').click()
```

```python
driver.find_element_by_xpath('/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/div/ul[2]/li[3]/a').click()
```

```python
repository=[]
description =[]
contributor = []
language_used =[]
```

```python
a = driver.find_elements_by_xpath("//h1[@class='h3 lh-condensed']/a")
for i in a:
    repository.append(i.text)
    
```

```python
url=[]
for i in driver.find_elements_by_xpath("//h1[@class='h3 lh-condensed']/a"):
    url.append(i.get_attribute("href"))
```

```python
for i in url:
    driver.get(i)
    language = []  
    
    
        
    languages = driver.find_elements_by_xpath("//li[@class='d-inline']//a//span[1]")
    if languages:
        for i in languages:
            language.append(i.text)
    else:
        language.append('no language used')
    language_used.append(language)
            
    
```

```python
for i in url:
    driver.get(i)
    time.sleep(2)
  
    try:
        c = driver.find_element_by_xpath("//h2[@class='h4 mb-3']/a[contains(text(),'Contributors')]/span")
        contributor.append(c.text)
    except NoSuchElementException:
        contributor.append('-')



    try:
        k = driver.find_element_by_xpath("//p[@class='f4 mt-3']")
        description.append(k.text)
    except NoSuchElementException:
        description.append('-')

```

```python
print(len(repository),len(description),len(contributor),len(language_used))
```

```python
df = pd.DataFrame({"Repository":repository,"Description":description,"Contributors count":contributor,"Language_used":language_used})
df
```

```python
driver.close()
```
